# Skill Challenge

```python
# players with nationality = english =
english_players = players.nationality == 'England'
english_players
```

```python
# market value is > 2 X Average market
players.market_value.sum()
```

```python
players.market_value.shape[0]
```

```python
average_mv = players.market_value.sum() / players.market_value.shape[0]
average_mv
```

```python
more_than_avg = players.market_value > (2 * average_mv)
```

```python
# have more than 400 views or a new signing
popular_xor = (players.page_views> 4000) ^ (players.new_signing >= 1)
```

```python
result = (english_players & more_than_avg & popular_xor)
```

```python
players.loc[result]
```
