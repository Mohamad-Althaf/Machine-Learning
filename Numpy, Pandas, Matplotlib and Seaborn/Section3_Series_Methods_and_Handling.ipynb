Filtering : filter() where(), and Mask()

```python
for i in alcohol_series.items():
  if i[0][0] == 'V':
    print(i)
```

```python
for i in alcohol_series.items():
  if 'stan' in i[0]:
    print(i)
```

```python
# Filter Function
alcohol_series.filter(regex = '^V')
```

```python
alcohol_series.filter(like = 'stan')
```

```python
alcohol_series[alcohol_series > 200]
```

```python
# Where() Method
```

```python
alcohol_series.where(lambda x: x>200).dropna()
```

```python
alcohol_series.where(lambda x: x<200).dropna()
```

```python
# Mask Method
```

```python
alcohol_series.mask(lambda x: x<200).dropna()
```
