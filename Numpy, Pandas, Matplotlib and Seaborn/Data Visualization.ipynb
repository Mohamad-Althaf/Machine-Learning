```python
import matplotlib.pyplot as plt
```

### Line Graph

Single plot in graph

```python
# x axis values
x = [1, 2, 3]
y = [2, 4, 1]

plt.plot(x, y)

plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('My First Graph ! ')

plt.grid(False, color = 'k')

plt.show()
```

Multiple Plot in graph

```python
# Line 1 points
x1 = [1, 2, 3]
y1 = [2, 4, 1]

plt.plot(x1, y1, label = 'Line 1')

# Line 2 points

x2 = [1, 2, 3]
y2 = [4, 2, 3]

plt.plot(x2, y2, label = 'Line 2')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title("Two lines in a graph")
plt.grid(False, color = 'k')
plt.legend()

plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 1, 5, 2, 6]

plt.plot(
    x, y, color = 'green',
          linestyle = 'dashed',
          linewidth = 5,
          marker = 'o',
          markerfacecolor = 'blue',
          markersize = 20
)

plt.ylim(1, 8)
plt.xlim(1, 8)

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title("Some cool customizations !")

plt.show()
```

## Bar Graph

```python
left = [1, 2, 3, 4, 5]
right = [10, 24, 4, 17, 9]

tick_label = ['one', 'two', 'three', 'four', 'five']

plt.bar(left, right, tick_label = tick_label,
        width = 0.5, color = ['red', 'green'])

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title('My bar chart !')

plt.show()
```

## Histogram

```python
ages = [10, 24, 4, 17, 9, 1, 2, 3, 4, 5]

range = (0, 20)
bins = 4

plt.hist(ages, bins, range, color = 'blue', histtype = 'bar', rwidth = 0.8)

plt.xlabel('age')
plt.ylabel('Frequency')

plt.title("My Histogram")

plt.show()
```

## Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 7, 5, 8, 9, 11, 13, 4]

plt.scatter(x, y, label = 'stars', color = 'green', marker = "*", s = 200)

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title("My Scatter Plot")

plt.legend()
plt.grid(False)
plt.show()
```

## Pie Chart

```python
activities = ['eat', 'sleep', 'work', 'play']

slices = [3, 5, 7, 6]

colors = ['g', 'r', 'b', 'm']

plt.pie(slices, labels = activities, colors = colors,
        startangle = 80, shadow = True, explode = (0.1, 0, 0.0, 0),
        radius = 1.2, autopct = '%1.1f%%')

plt.gca().legend(loc = 'center left', bbox_to_anchor = (1, 0.5))

plt.show()
```

## Sinusoidal wave

```python
import numpy as np
x = np.arange(0, 2 * (np.pi), 0.1)

y = np.sin(x)

plt.plot(x, y)
plt.grid(True)
plt.show()
```

## Creating multiple plots in a graph

```python
def create_plot(ptype):
  x = np.arange(-10, 10, 0.01)
  if ptype == 'linear':
    y = x
  elif ptype == 'quadratic':
    y = x ** 2
  elif ptype == 'cubic':
    y = x ** 3
  elif ptype == 'quartic':
    y = x ** 4

  return (x, y)

plt.style.use('fivethirtyeight')

fig = plt.figure()

plt1 = fig.add_subplot(2, 2, 1)
plt2 = fig.add_subplot(2, 2, 2)
plt3 = fig.add_subplot(2, 2, 3)
plt4 = fig.add_subplot(2, 2, 4)


x, y = create_plot('linear')
plt1.plot(x, y, color = 'r')
plt1.set_title('$y_1 = x$')

x, y = create_plot('quadratic')
plt2.plot(x, y, color = 'b')
plt2.set_title('$y_2 = x^2$')

x, y = create_plot('cubic')
plt3.plot(x, y, color = 'g')
plt3.set_title('$y_3 = x^3$')

x, y = create_plot('quartic')
plt4.plot(x, y, color = 'k')
plt4.set_title('$y_4 = x^4$')

fig.subplots_adjust(hspace = 1, wspace = 1)
plt.show()
```

```python
## Using Gridsize

gridsize = (3, 2)
fig = plt.figure(figsize = (12, 8))
ax1 = plt.subplot2grid(gridsize, (0, 0), colspan = 1, rowspan = 1)
ax2 = plt.subplot2grid(gridsize, (1, 0), colspan = 1, rowspan = 1)
ax1 = plt.subplot2grid(gridsize, (2, 0), colspan = 1, rowspan = 1)
plt.show()
```

### Stack Plot

```python
rng = np.arange(50)
rnd = np.random.randint(0, 10, size = (3, rng.size))
yrs = 1950 + rng
fig, ax = plt.subplots(figsize = (5, 3))
ax.stackplot(yrs, rng + rnd , labels = ['A', 'B', 'C'])
ax.set_title("Combined debt growth over time")
ax.set_ylabel('Total debt')
ax.set_xlim(left = yrs[0], right = yrs[-1])
fig.tight_layout()
```

### Combination of multiple plots

```python
x = np.random.randint(low = 1, high = 11, size = 50)
y = x + np.random.randint(1, 5, size = x.size)
data = np.column_stack((x, y))

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (8, 4))

ax1.scatter(x = x, y = y, marker = 'o', edgecolor = 'b')
ax1.set_title('Scatter : $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

ax2.hist(data , bins = np.arange(data.min(), data.max()), label = ('x', 'y'))
ax2.legend(loc = (0.65, 0.8))
ax2.set_title("Frequencies of $x$ and $y$")
```
