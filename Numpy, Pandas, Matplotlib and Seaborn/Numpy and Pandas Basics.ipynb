### Data Preparation

```python
import numpy as np
import pandas as pd
```

```python
age = np.random.choice([15, 20, 30, 45, 12, '10', 15, '34', 7, 'missing'], 50)
fico = np.random.choice(['100-150', '150-200', '200-250', '250-300'], 50)
city = np.random.choice(['Mumbai', 'Delhi', 'Chennai', 'Kolkata'], 50)
ID = np.arange(50)
rating = np.random.choice(['Excellent', 'Good', 'Bad', 'Pathetic'], 50)
balance = np.random.choice([10000, 20000, 30000, 40000, np.nan, 50000, 60000], 50)
children = np.random.randint(high = 5, low = 0, size = (50, ))
```

```python
mydata = pd.DataFrame(
    {
        'ID' : ID,
        'age' : age,
        'fico' : fico,
        'city' : city,
        'rating' : rating,
        'balance' : balance,
        'children' : children
    }
)
```

```python
mydata.head(10)
```

```python
mydata.dtypes
```

```python
## Converting the age columns' datatype to int
mydata.age.value_counts()
```

```python
"""
errors = 'coerce' -->  The age col is object, we are trying to convert into integer.
But in the age col we have string values which has to be handled separately.
Hence we use  errors = 'coerce' , so that it will convert the string values to NaN.
"""

mydata['age'] = pd.to_numeric(mydata['age'], errors = 'coerce')
```

```python
mydata.dtypes
```

```python
mydata.age.value_counts()
```

```python
mydata['age'].isnull().sum()
```

##### Filling NaN - Method 1

```python
mydata[mydata['age'].isnull()].age
```

```python
mydata['age'].fillna(mydata['age'].mean())
```

##### Filling NaN - Method 2

```python
mydata['age'].isnull().sum()
```

```python
mydata.loc[mydata['age'].isnull(), 'age'] = mydata['age'].mean()

mydata['age_children_ratio'] = mydata['age'] / mydata['children']
```

```python
mydata
```

##### Adding columns

```python
mydata['const_var'] = 100
mydata['balance_log'] = np.log(mydata['balance'])
```
