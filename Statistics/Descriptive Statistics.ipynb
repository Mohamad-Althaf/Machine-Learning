#**Descriptive Statistics**

```python
import random
import pandas as pd
import numpy as np
```

EXAMPLE 1

```python
random_list = []

for i in range(10, 50):
  n=random.randint(1,200)
  random_list.append(n)

print(random_list)
```

```python
df = pd.DataFrame(random_list, columns = ['Numbers'])
```

```python
df
```

```python
df.Numbers.plot()
```

```python
df.Numbers.hist()
```

```python
df.Numbers.plot.density()
```

### Calculating the MEAN, MEDIAN, MODE, STANDARD DEVIATION, VARIANCE

```python
print(df.Numbers.min())
print(df.Numbers.max())
print(df.Numbers.mean())
print(df.Numbers.median())
print(df.Numbers.mode())
print(df.Numbers.std())
print(df.Numbers.var())
```

```python
print(df.Numbers.skew())
print(df.Numbers.kurt())
```

EXAMPLE 2

```python
score = [10, 15, 20, 17, 24, 27, 19, 32, 61]
freq = [5, 24, 12, 12, 6, 18, 21, 66, 4]
```

```python
total_score = []

for i in range(0, len(score)):
  total_score.append(score[i] * freq[i])
```

```python
print(total_score)
```

```python
number_of_students = sum(freq)
print(number_of_students)
```

```python
## Mean

Mean = sum(total_score) / number_of_students
print(Mean)
```
