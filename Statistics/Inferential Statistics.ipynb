Z Test

```python
from statsmodels.stats.weightstats import ztest
z_val , p_val = ztest(data[data['Neighborhood']=='Edwards']['SalePrice'],
                      value = data['SalePrice'].mean())
```

```python
print("Z_Value : ", z_val , " | P_Value : ", p_val)
```

```python
z_val , p_val = ztest(data[data['Neighborhood']=='OldTown']['SalePrice'],
                      value = data['SalePrice'].mean())
print("Z_Value : ", z_val , " | P_Value : ", p_val)
```

```python
mean_ = data[data['Neighborhood']=='Edwards']['SalePrice'].mean()
stdev = data[data['Neighborhood']=='Edwards']['SalePrice'].std()

from scipy import stats

z_score = (200000 - mean_) / stdev
stats.norm.cdf(z_score)
```

```python
mean_ = data[data['Neighborhood']=='IDOTRR']['SalePrice'].mean()
stdev = data[data['Neighborhood']=='IDOTRR']['SalePrice'].std()

from scipy import stats

z_score = (200000 - mean_) / stdev
stats.norm.cdf(z_score)
```

T Test

```python
t_val , p_val = stats.ttest_1samp(data[data['Neighborhood']=='Edwards']['SalePrice'].sample(60),
                                  popmean = data['SalePrice'].mean())

print("T_Value : ", t_val , " | P_Value : ", p_val)
```

```python
t_val , p_val = stats.ttest_1samp(data[data['Neighborhood']=='IDOTRR']['SalePrice'].sample(20),
                                  popmean = data['SalePrice'].mean())

print("T_Value : ", t_val , " | P_Value : ", p_val)
```

```python
a = data[data['YearBuilt'] <= 1960]['SalePrice']
b = data[data['YearBuilt'] > 1999]['SalePrice']
```

```python
stats.ttest_ind(a, b, axis = 0, equal_var = True)
```

Chi Square Test

```python
def compute_freq_chi(x, y):
  freqtab = pd.crosstab(x, y)
  print("=== Frequency Table ===")
  print(freqtab)
  print("=======================")
  chi2 , pval, dof, expected = stats.chi2_contingency(freqtab)
  print("ChiSquare test statistics : ", chi2)
  print("P_Value : ", pval)
  return

price = pd.qcut(data['SalePrice'], 3, labels = ['High', 'Medium', 'Low'])
compute_freq_chi(data.LandContour, price)
```
